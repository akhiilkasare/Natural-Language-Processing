```python
# Tokenization basically converts the original text data into tokens 
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
mystring = '"We\'re moving to L.A!"'
```

```python
mystring
```

```python
doc = nlp(mystring)
```

```python
for token in doc:
    print(token.text)
```

# Prefixes Suffixes and Infixes
Spacy removes commas,quotation marks,punctuation at the end of the sentences.But it does not removes punctuation as a part of email address, website or numeric value

```python
doc2 = nlp(u"We're here to help! Send snail-mail, email support@oursite.com or visit us at http://www.oursite.com!")

for t in doc2:
    print(t)
```

```python
doc3 = nlp(u'A 5km NYC cab ride costs $10.30')
```

```python
for t in doc3:
    print(t)
```
